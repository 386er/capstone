```python
import seaborn as sns
import matplotlib.pyplot as plt
import os
import warnings
import pandas as pd
from matplotlib import gridspec
import helperModule as hm

sns.set(context="paper", font="monospace")

warnings.filterwarnings("ignore")
%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```python
data_folder =  '.' + os.sep + 'data' + os.sep

fields = pd.read_csv(data_folder + 'Field Names.csv', header=None)
field_names = list(fields[0])
data = pd.read_csv(data_folder + 'KDD99Train.csv', names=field_names)

warnings.filterwarnings("ignore")
```

```python
#########################
##### Correlation Heatmap
#########################

corrmat = data.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True)
f.savefig('corrheat.pdf')
```
